```python
# Find Out The State Which Has Minimum Total Cases?
data.sort_values(by='Total Cases',ascending=True)[0:1]
```

```python
# Second Way
data.loc[data['Total Cases'].idxmin()] # idxmin --- return index of min. value.
```

```python
# Find Out The State Which Has Maximum Total Cases?
data.sort_values(by='Total Cases',ascending=False).head(1)
```

```python
# second method
data.loc[data['Total Cases'].idxmax()] # idxmin -- return index of max. value.
```

```python
# Display All States Of India From DataFrame?
data['State/UTs']
```

```python
# Display All The Data Of States Having Deaths Greater Than 10000?
data[(data['Deaths']>10000)]
```

```python
# Count The Total Cases Of India?
data['Total Cases'].sum()
```

```python
# What is the Death Ratio Of Maharashtra?
data['Death Ratio'][20]
```

```python
# Which State Has Minimum Death Ratio?
data.groupby('Death Ratio').min().head(1)
```

```python
# Find Out The Discharged Patients From The States Maharashtra,Karnataka,MP,Gujrat?
print('Discharge Patients of Maharastra',data['Discharged'][20])
print('Discharge Patients of Karnataka',data['Discharged'][15])
print('Discharge Patients of MP',data['Discharged'][19])
print('Discharge Patients of Gujrat',data['Discharged'][10])
```

```python
# Find Out The States Having Total Cases between 1 lac to 5 lac?
data[(data['Total Cases']>100000) & (data['Total Cases']<500000)]
```

```python
# second method
data[data['Total Cases'].between(100000,500000)]
```

```python
# Display All The Data Of The States Which Has Active Cases Less Than 1000?
data[data['Active']<1000]
```

```python
# second way to solve the minimum Death ratio 
data.sort_values(by='Death Ratio',ascending=True).head(1) 
```

# Make a seprate CSV file of data of the statewhich has active cases less than 1000

```python
data[data['Active']<1000].to_csv('newdata.csv')
```

```python
# Total no of death in India
print('Total no of death in India',data['Deaths'].sum())
```

# Top 10 highest Population States in india

```python
data.loc[data.sort_values('Population',ascending=True).index.tolist(),['State/UTs','Population']].head(10)
```

```python
# Second way
data.iloc[data.sort_values('Population',ascending=True).index.tolist(),[0,8]].head(10)
```
